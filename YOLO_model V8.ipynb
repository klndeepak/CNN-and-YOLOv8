```python
from ultralytics import YOLO

# Load a model
model = YOLO("yolov8n.yaml")  # build a new model from scratch
model = YOLO("yolov8n.pt")  # load a pretrained model (recommended for training)

# Use the model
model.train(data="coco128.yaml", epochs=3)  # train the model
metrics = model.val()  # evaluate model performance on the validation set
results = model("https://ultralytics.com/images/bus.jpg")  # predict on an image
path = model.export(format="onnx")  # export the model to ONNX format
```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

image = results[0].orig_img
boxes = results[0].boxes.xyxy.tolist()

plt.imshow(image)

for box in boxes:
    x, y, w, h = box[:4]  # Extract the first four values (x, y, width, height) from the box
    rect = patches.Rectangle((x, y), w, h, linewidth=1, edgecolor='r', facecolor='none')
    #edgecolor and facecolor are for the bounding boxes. We are creating transparent box with red color edges.
    rect.set_linewidth(4)
    #By using `plt.gca().add_patch(rect)`, the rectangle patch is added to the current axes, which is the Axes instance returned by `plt.gca()`. This allows the rectangle to be displayed on the image plot.

    plt.gca().add_patch(rect)

plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

image = results[0].orig_img
boxes = results[0].boxes.xyxy.tolist()
labels = results[0].names

# Define color map for each class
class_colors = {
    'person': 'red',
    'car': 'blue',
    'dog': 'green',
    # Add more classes and colors as needed
}

plt.imshow(image)

for box, label in zip(boxes, labels):
    x, y, w, h = box[:4]  # Extract the first four values (x, y, width, height) from the box
    class_name = str(label)  # Convert the label to an integer
    class_color = class_colors.get(class_name, 'yellow')  # Get the color for the class, default to yellow if not found

    rect = patches.Rectangle((x, y), w, h, linewidth=1, edgecolor=class_color, facecolor='none')
    rect.set_linewidth(4)
    plt.gca().add_patch(rect)

    # Display class name
    plt.text(x, y, class_name, color=class_color, backgroundcolor='w', fontsize=8)

plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

image = results[0].orig_img
boxes = results[0].boxes.xyxy.tolist()
labels = results[0].names

# Define a list of colors for bounding boxes
box_colors = ['blue', 'green', 'yellow', 'orange']  # Add more colors as needed

plt.imshow(image)

for i, (box, label) in enumerate(zip(boxes, labels)):
    x, y, w, h = box[:4]  # Extract the first four values (x, y, width, height) from the box
    rect = patches.Rectangle((x, y), w, h, linewidth=1, edgecolor=box_colors[i % len(box_colors)], facecolor='none')
    rect.set_linewidth(4)
    plt.gca().add_patch(rect)

    # Display class name
    plt.text(x, y, label, color=box_colors[i % len(box_colors)], backgroundcolor='w', fontsize=8)

plt.show()
```
